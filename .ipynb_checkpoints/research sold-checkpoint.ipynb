```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler,Normalizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
sample=pd.read_csv('sample_submission.csv')
train
```

```python
train['date']=pd.to_datetime(train['date'])
train["year"]=train["date"].dt.year
train["month"]=train["date"].dt.month
train["day"]=train["date"].dt.day
grp_month_day=train.groupby(["month","day"])

train
```

```python
def func(x):
    scalar=StandardScaler()
    scalar.fit(x.values.reshape(-1,1))
    x=scalar.transform(x.values.reshape(-1,1)).reshape(x.values.size)+5
    return x
# def func(x):
#     nor=Normalizer()
#     x=nor.transform(x.values.reshape(-1,1)).reshape(x.values.size)
#     return x

def func2(x):
    print(x)

scaler = StandardScaler()
grp_date=train.groupby("date")
grp_month=train.groupby(["year","month"])
train["nor_sold"]= grp_month["num_sold"].transform(func)
grp_all=train.groupby(["country","store","product","month","day"])
train["days_year_avg"]=grp_all["nor_sold"].transform("mean")  
grp_month_day.get_group((3,29))
```

```python
train_data=train.drop(["date","nor_sold","id"],axis=1)
```

```python

def factorize_dataframe(df, columns_to_factorize):
    mappings = {}
    
    for col in columns_to_factorize:
        labels, unique_values = pd.factorize(df[col])
        df[col] = labels+1
        mappings[col] = dict(zip(unique_values, labels+1))
    df["year"]=df["year"]-2000
    return df, mappings




# Columns to factorize
columns_to_factorize = ['country', 'store', 'product']

# Apply the function and get the transformed DataFrame and mappings
train_data_factorize, mappings = factorize_dataframe(train_data, columns_to_factorize)

# print("Transformed DataFrame:")
# print(train_data_factorize)
print("\nMappings:")
print(mappings)
train_data_factorize
```

```python
Y=train_data_factorize.loc[:,["num_sold"]]
X=train_data_factorize.drop(["num_sold","country","store","product"],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Step 3: Create and fit the Ridge regression model
alpha = 1.0  # Regularization strength (lambda)
ridge_model = Ridge(alpha=alpha)
ridge_model.fit(X_train, y_train)

# Step 4: Make predictions on the test set
y_pred = ridge_model.predict(X_test)

# Step 5: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Step 3: Make predictions on the test set
y_pred = linear_model.predict(X_test)

# Step 4: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Step 5: Get the coefficients and intercept of the linear model
coefficients = linear_model.coef_
intercept = linear_model.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
degree = 2  # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

# Fit a linear regression model to the polynomial features
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_poly)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
